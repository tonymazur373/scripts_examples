```python
filename = "IntegrioUPDATEArtificialIntelligence_AI_inRea"
```

```python
text = open('C:/Users/TonyM/Downloads/'+ filename +'.html', encoding="utf8").read()
```

```python
head = text.split("</head>")[0]
```

```python
body = text.split("</head>")[1]
```

```python
background_color_classes = [x for x in head.split("}.") if "background-color" in x]
```

```python
background_color_classes_dict = {}
for x in background_color_classes:
    class_name = x.split("{")[0]
    color_code = x.split("background-color:")[1][:7]
    if color_code != "#ffffff": # #ffffff is white
        background_color_classes_dict[class_name] = color_code
```

```python
background_color_classes_dict
```

```python
background_color_classes_names = list(background_color_classes_dict.keys())
```

```python
import re
```

```python
background_color_classes_names
```

```python
highlights = [row for row in re.findall(r'<span.+?/span>', body)
                         for name in background_color_classes_names 
                         if re.search(name+r'\b', row) is not None]
```

```python
total_words = 0
preprocessed_highlights = []
for words in highlights:
    text_1 = re.sub(r'<.+?>', '', words) #removes sequences between < and > like </span>, </a> , <a class="c12" href=...>
    text_2 = re.sub('&'+r'\S+'+';', '', text_1) # removes special characters like &rsquo; &amp; &nbsp;
    text_3 = re.sub('-', '', text_2) #remove dashes for they mess with word counting - for r'\w+'-pattern considers parts separated by a dash as separate words
    preprocessed_highlights.append(text_3)
    total_words = total_words + len(re.findall(r'\w+', text_3)) #counting total number of words
```
